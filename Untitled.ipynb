# Análise de padrões de mobilidade

```python
bikes['day_of_week'] = bikes['departure'].dt.day_name()

bikes['hour'] = bikes['departure'].dt.hour
```

```python
peak_time = bikes.groupby(['day_of_week', 'hour']).size().reset_index(name = 'count')
```

```python
top_peak_time = peak_time.sort_values(by = 'count', ascending = False).drop_duplicates(subset = 'day_of_week')
```

```python
top_peak_time['day_hour'] = top_peak_time['day_of_week'] + '-' + top_peak_time['hour'].astype(str)
```

```python
plt.figure(figsize = (8, 6))

sns.barplot(x = 'day_hour', y = 'count', data = top_peak_time, palette = 'viridis')

plt.title('Horários de maior pico em diferentes dias da semana')

plt.xlabel('Dia-Hora')
plt.ylabel('Contagem')

plt.xticks(rotation = 45, ha = 'right')

for i, v in enumerate(top_peak_time['count'].values):
    plt.text(i, v, str(v), ha = 'center', va = 'bottom')

plt.show()
```

```python
bottom_peak_time = peak_time.sort_values(by = 'count').drop_duplicates(subset = 'day_of_week')
```

```python
bottom_peak_time['day_hour'] = bottom_peak_time['day_of_week'] + '-' + bottom_peak_time['hour'].astype(str)
```

```python
plt.figure(figsize = (8, 6))

sns.barplot(x = 'day_hour', y = 'count', data = bottom_peak_time, palette = 'viridis')

plt.title('Horários de menor pico em diferentes dias da semana')

plt.xlabel('Dia-Hora')
plt.ylabel('Contagem')

plt.xticks(rotation = 45, ha = 'right')

for i, v in enumerate(bottom_peak_time['count'].values):
    plt.text(i, v, str(v), ha = 'center', va = 'bottom')

plt.show()
```

Vemos que o horário de pico onde as pessoas pegam as bicicletas para dar uma volta é em torno de 15h e 17h que talvez seja o horário onde termina o espediente de pessoas que trabalham, ou o horário do termino das aulas de jovens. Já nos horários de menor pico varia entre 3h e 6h um horário bem inacessível para a maioria das pessoas.

```python
bikes['month'] = bikes['departure'].dt.month_name()
```

```python
bikes['month'].value_counts()
```

```python
seasons = {'January': 'Winter', 'February': 'Winter', 'March': 'Spring',
           'April': 'Spring', 'May': 'Spring', 'June': 'Summer',
           'July': 'Summer', 'August': 'Summer', 'September': 'Fall',
           'October': 'Fall', 'November': 'Fall', 'December': 'Winter'}
```

```python
bikes['season'] = bikes['month'].map(seasons)
```

```python
peak_season = bikes.groupby('season').size().reset_index(name = 'count')
```

```python
plt.figure(figsize = (8, 6))

sns.barplot(x = 'season', y = 'count', data = peak_season, palette = 'husl')

plt.title('O pico de uso de bicicletas nas estações do ano')

plt.xlabel('Estação')
plt.ylabel('Contagem')

for i, v in enumerate(peak_season['count'].values):
    plt.text(i, v, str(v), ha = 'center', va = 'bottom')

plt.show()
```

Como era de se esperar, as pessoas tendem a usar mais as bicicletas no verão, quando o clima é ideal para atividades ao ar livre. No entanto, durante o inverno, o uso das bicicletas cai drasticamente, chegando a ponto de não haver nenhum registro de rotas.

```python
top_departure_season = bikes.groupby(['season', 'departure_name']).size().reset_index(name = 'count')
```

```python
top_10_departures_season = top_departure_season.groupby('season').apply(lambda x: x.nlargest(10, 'count')).reset_index(drop=True)
```

```python
top_10_departures_Fall = top_10_departures_season[top_10_departures_season['season'] == 'Fall']

top_10_departures_Spring = top_10_departures_season[top_10_departures_season['season'] == 'Spring']

top_10_departures_Summer = top_10_departures_season[top_10_departures_season['season'] == 'Summer']
```

```python
seasons_data = [top_10_departures_Fall, top_10_departures_Spring, top_10_departures_Summer]
```

```python
fig, axs = plt.subplots(3, 1, figsize = (12, 14))

for i, data in enumerate(seasons_data):
    
    sns.barplot(x = 'count', y = 'departure_name', data = data, palette = 'viridis', ax = axs[i])
    axs[i].set_title(f'Locais com maior número de partidas de acordo com a estação {data.season.values[0]}')

plt.show()
```

Observamos as áreas com maior concentração de uso de bicicletas ao longo das diferentes estações do ano. Os pontos de partida Itämerentori, Töölönlahdenkatu e Kamppi (M) se destacam como os locais com maior fluxo de usuários.

```python
bins = [-6, 3, 12, 21, 30, 33]

labels = ['-6 a 3', '3 a 12', '12 a 21', '21 a 30', '30 a 33']
```

```python
bikes['temperature_interval'] = pd.cut(bikes['Air temperature (degC)'], bins = bins, labels = labels, right = False)
```

```python
peak_temperature = bikes.groupby('temperature_interval').size().reset_index(name = 'count')
```

```python
plt.figure(figsize = (8, 6))

sns.barplot(x = 'temperature_interval', y = 'count', data = peak_temperature, palette = 'viridis')

plt.title('O impacto da temperatura no uso de bicicletas')

plt.xlabel('Temperatura em Celsius')
plt.ylabel('Contagem')

for i, v in enumerate(peak_temperature['count'].values):
    plt.text(i, v, str(v), ha = 'center', va = 'bottom')

plt.show()
```

```python
bikes['duration'] = bikes['return'] - bikes['departure']
```

```python
bikes['duration'] = (bikes['duration'].dt.total_seconds() / 60).round(2)
```

```python
temperature_vs_time = bikes.groupby('temperature_interval')['duration'].mean().reset_index(name = 'mean_minutes')
```

```python
plt.figure(figsize = (8, 6))

sns.barplot(x = 'temperature_interval', y = 'mean_minutes', data = temperature_vs_time, palette = 'viridis')

plt.title('O impacto da temperatura no tempo de uso de bicicletas')

plt.xlabel('Temperatura em Celsius')
plt.ylabel('Minutos')

for i, v in enumerate(temperature_vs_time['mean_minutes'].values):
    plt.text(i, v, str(v.round(2)), ha = 'center', va = 'bottom')

plt.show()
```

Agora podemos ver que os usuários costumam usar bicicletas com mais frequência quando a temperatura está entre 12°C a 21°C. No entanto, o tempo médio de uso das bicicletas é maior em temperaturas mais baixas, especificamente entre -6°C e 3°C. Isso sugere que apesar de uma preferência geral por climas moderados, os usuários que pedalam em temperaturas frias parecem realizar trajetos mais longos ou permanecer com as bicicletas por períodos mais extensos.

# Previsão da demanda por estruturas de mobilidade

```python
season_departure_bike_demand = bikes.groupby(['season', 'departure_name']).size().reset_index(name = 'count').sort_values(by = 'count', ascending = False)
```

```python
top10_summer_departure_bike_demand = season_departure_bike_demand[season_departure_bike_demand['season'] == 'Summer'][:10]

top10_spring_departure_bike_demand = season_departure_bike_demand[season_departure_bike_demand['season'] == 'Spring'][:10]

top10_fall_departure_bike_demand = season_departure_bike_demand[season_departure_bike_demand['season'] == 'Fall'][:10]
```

```python
datasets = [top10_summer_departure_bike_demand, top10_spring_departure_bike_demand, top10_fall_departure_bike_demand]
```

```python
fig, axes = plt.subplots(3, 1, figsize = (8, 16))

for i, data in enumerate(datasets):
    
    sns.barplot(x = 'count', y = 'departure_name', data = data, palette = 'viridis', ax = axes[i])
    axes[i].set_title(f'Top 10 locais de partida com maior demanda de bicicletas na estação {data.season.values[0]}', fontsize = 16)
    axes[i].set_ylabel('')
    axes[i].set_xlabel('Contagem')

plt.show()
```

```python
season_return_bike_demand = bikes.groupby(['season', 'return_name']).size().reset_index(name = 'count').sort_values(by = 'count', ascending = False)
```

```python
top10_summer_return_bike_demand = season_return_bike_demand[season_return_bike_demand['season'] == 'Summer'][:10]

top10_spring_return_bike_demand = season_return_bike_demand[season_return_bike_demand['season'] == 'Spring'][:10]

top10_fall_return_bike_demand = season_return_bike_demand[season_return_bike_demand['season'] == 'Fall'][:10]
```

```python
datasets = [top10_summer_return_bike_demand, top10_spring_return_bike_demand, top10_fall_return_bike_demand]
```

```python
fig, axes = plt.subplots(3, 1, figsize = (8, 16))

for i, data in enumerate(datasets):
    
    sns.barplot(x = 'count', y = 'return_name', data = data, palette = 'viridis', ax = axes[i])
    axes[i].set_title(f'Top 10 locais de retorno com maior demanda de bicicletas na estação {data.season.values[0]}', fontsize = 16)
    axes[i].set_ylabel('')
    axes[i].set_xlabel('Contagem')

plt.show()
```

Conseguimos identificar os locais que requerem maior atenção em relação à demanda por bicicletas. Isso permite direcionar recursos de forma estratégica, garantindo que as necessidades específicas de cada estação sejam atendidas de maneira eficiente

# Análise de eficácia e segurança de ciclovias

```python
departure_return_speed = bikes.groupby(['departure_name', 'return_name'])['avg_speed (km/h)'].mean().reset_index().sort_values(by = 'avg_speed (km/h)', ascending = False)
```

```python
departure_return_speed = departure_return_speed[departure_return_speed['avg_speed (km/h)'] >= 0]
```

```python
departure_return_speed['route'] = departure_return_speed['departure_name'] + '->' + departure_return_speed['return_name']
```

```python
departure_return_speed = departure_return_speed.head(10)
```

```python
plt.figure(figsize = (8, 6))

sns.barplot(x = 'avg_speed (km/h)', y = 'route', data = departure_return_speed, palette = 'viridis')

plt.title('Rotas onde os ciclistas tendem a aumentar a velocidade')

plt.xlabel('Velocidade média (km/h)')
plt.ylabel('Rotas')

for i, v in enumerate(departure_return_speed['avg_speed (km/h)'].values):
    plt.text(v, i, str(v.round(2)), ha = 'left', va = 'center')

plt.show()
```

```python
bikes.head()
```

```python
departure_return_duration = bikes.groupby(['departure_name', 'return_name'])['duration (min.)'].mean().reset_index().sort_values(by = 'duration (min.)', ascending = False)
```

```python
departure_return_duration = departure_return_duration[departure_return_duration['duration (min.)'] > 0]
```

```python
departure_return_duration['route'] = departure_return_duration['departure_name'] + '->' + departure_return_duration['return_name']
```

```python
plt.figure(figsize = (8, 6))

sns.barplot(x = 'duration (min.)', y = 'route', data = departure_return_duration.head(10), palette = 'viridis')

plt.title('Rotas onde a duração do percurso é mais longo')

plt.xlabel('Duração (minutos)')
plt.ylabel('Rotas')

for i, v in enumerate(departure_return_duration['duration (min.)'].head(10).values):
    plt.text(v, i, str(v.round(2)), ha = 'left', va = 'center')

plt.show()
```

```python
plt.figure(figsize = (8, 6))

sns.barplot(x = 'duration (min.)', y = 'route', data = departure_return_duration.tail(10), palette = 'viridis')

plt.title('Rotas onde a duração do percurso é mais curto')

plt.xlabel('Duração (minutos)')
plt.ylabel('Rotas')

for i, v in enumerate(departure_return_duration['duration (min.)'].tail(10).values):
    plt.text(v, i, str(v.round(2)), ha = 'left', va = 'center')

plt.show()
```

Com estes gráficos o departamento de trânsito pode usar esses dados para identificar áreas onde a infraestrutura pode ser aprimorada, como aumento de segurança em cruzamentos perigosos, ou criação de ciclovias onde há uma demanda reprimida.
